### Reconhecimento Facial

```python
from PIL import Image
import numpy as np
import os
import cv2
```

```python
imagem = 'Arquivos/yalefaces/train/'
```

```python
def dados_imagem():
    caminhos = [os.path.join('Arquivos/yalefaces/train/', f) for f in os.listdir('Arquivos/yalefaces/train/')]
    faces = []
    ids= []
    for caminho in caminhos:
        imagem = Image.open(caminho).convert('L')
        imagem_np = np.array(imagem,'uint8')
        id = int(os.path.split(caminho)[1].split('.')[0].replace('subject',''))
        ids.append(id)
        faces.append(imagem_np)
    return np.array(ids), faces
```

```python
ids, faces = dados_imagem()
```

```python
print(ids)
```

```python
print(faces)
```

```python
lbph = cv2.face.LBPHFaceRecognizer_create()
lbph.train(faces,ids)
lbph.write('classificadorLBPG.yml')
```

#### Classificao

```python
reconhecedor = cv2.face.LBPHFaceRecognizer_create()
reconhecedor.read('classificadorLBPG.yml')
```

```python
imagem_teste = 'Arquivos/yalefaces/test/subject10.sad.gif'
```

```python
imagem = Image.open(imagem_teste).convert('L')
imagem_np = np.array(imagem,'uint8')
print(imagem_np)
```

```python
idprevisto, _ = reconhecedor.predict(imagem_np)
idprevisto
```
